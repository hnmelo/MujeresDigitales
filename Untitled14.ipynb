```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
# Escriba su código aquí
df = pd.read_csv('accidents_sample.csv', sep=';', parse_dates=[['DATE', 'TIME']])
df.head(3)
```

```python
df['CONTEO'] = df['DATE_TIME'].dt.to_period('D')
df['HORA'] = df['DATE_TIME'].dt.hour
```

```python
mask = df[df['CONTEO'] < '01/02/2018']
group = mask.groupby('HORA').count()
group
```

```python
sns.lineplot(x = group.index, y = 'ANO-DIA', data = group)
plt.xticks(rotation = 60)
plt.ylabel = 'CONTEO'
```
